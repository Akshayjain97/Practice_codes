```python
import numpy as np
import cv2

#read image
img_src = cv2.imread('/home/leno/cifar/cifar10/test/9654.jpg')

#edge detection filter
kernel = np.array([[0.0, -1.0, 0.0], 
                   [-1.0, 5.0, -1.0],
                   [0.0, -1.0, 0.0]])

kernel = kernel/(10*(np.sum(kernel)) if np.sum(kernel)!=0 else 1)

#filter the source image
img_rst = cv2.filter2D(img_src,-1,kernel)

#save result image
cv2.imwrite('/home/leno/cifar/cifar10/result.jpg',img_rst)
```
