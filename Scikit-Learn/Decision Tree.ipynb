```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("Data.csv",names=['s.n ','Price','Maintenance','capacity','aribag','profit'])
```

```python
def entropy(Col,final):
  num =[]
  Hs=[]
  distinct,count = np.unique(dataset[Col],return_counts=True)
  if Col==final:
    for i in range(len(distinct)):
      num.append((count[i]/np.sum(count))*np.log2(count[i]/(np.sum(count))))
    entpy = -np.sum(num)
    return entpy
  else:
    for i in range(len(distinct)):
      num.clear()
      df = dataset.loc[dataset[Col]==distinct[i]]
      distinct1,count1 = np.unique(df[final],return_counts=True)
      for j in range(len(distinct1)):
        num.append((count1[j]/np.sum(count1))*np.log2(count1[j]/(np.sum(count1))))
      Hs.append(np.sum(num))

  return Hs   
```

```python
def InfoG(final):
  node_ent = entropy(final,final)
  col = {'Price','Maintenance','capacity','aribag'}  
  node_ig={}
  categ_ent = []
  for i in col:
    distinct,count = np.unique(dataset[i],return_counts=True)
    for j in range(len(distinct)):
      categ_ent.append((count[j]/np.sum(count))*entropy(i,final)[j])
    final_cate = np.sum(categ_ent)
    node_ig[i]= node_ent + final_cate
    categ_ent.clear()

  global maxkey 
  maxkey = max(zip(node_ig.values(), node_ig.keys()))[1]
  print("highest info gain for :",maxkey)
  return node_ig
```
