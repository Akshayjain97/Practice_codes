```python
import numpy 
import pandas as pd	
```

```python
def sigmoid(inpt):
    return 1.0 / (1 + numpy.exp(-1 * inpt))
```

```python
data = pd.read_csv('Social_Network_Ads.csv')
input = data.iloc[:,:-1].values	
output= data.iloc[:,-1].values	
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
input[:, :] = sc.fit_transform(input[:, :-1])
input
```

```python
layer1 = 3
weight1 = numpy.random.uniform(low=-0.1, high=0.1, size=(input.shape[1], layer1))
layer2 = 2
weight2 = numpy.random.uniform(low=-0.1, high=0.1, size=(layer1, layer2))
output = 2
weight3 = numpy.random.uniform(low=-0.1, high=0.1, size=(layer2, output))
```

```python
a=input[0, :]
b=weight1
l1output = numpy.matmul(a,b)
l1outputs = sigmoid(l1output)
a=l1outputs
b=weight2
l2outputs = numpy.matmul(a,b)
l2utputs = sigmoid(l2outputs)
a=l2outputs
b=weight3
out_otuputs = numpy.matmul(a,b)
```

```python
predicted_label = numpy.where(out_otuputs == numpy.max(out_otuputs))[0][0]
print("Predicted class : ", predicted_label)
```
