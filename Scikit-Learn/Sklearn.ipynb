```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('c1.csv')
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
from sklearn import linear_model
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge
regressor2 = PolynomialFeatures(degree = 5)
x_poly = regressor2.fit_transform(x)
regressor3 = linear_model.Ridge(alpha=10)
regressor3.fit(x_poly,y)
```

```python
plt.scatter(x,y,color = 'red')
plt.plot(x,regressor3.predict(x_poly),color= 'blue')
plt.title('Poly')
plt.xlabel('X')
plt.ylabel('t|y')
plt.show()
```
