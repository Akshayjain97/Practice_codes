# SQL’s MIN(), MAX(), COUNT(), AVG(), and SUM() 

```python
print(df['SALARY'].min())
print(df['SALARY'].max())
print(df['SALARY'].sum())
print(df['SALARY'].mean())
print(df['SALARY'].count())
print(df['SALARY'].sum()/df['SALARY'].count())
```

# WHERE condition

```python
df.loc[(df['SALARY']>6000) & (df['SALARY']<10000),['EMPLOYEE_ID' ,'FIRST_NAME','SALARY']].drop_duplicates()
```

# ORDER BY

```python
df[['EMPLOYEE_ID' ,'FIRST_NAME','SALARY']].sort_values(['FIRST_NAME','SALARY'],ascending=False)
```

```python
df.loc[:,['EMPLOYEE_ID' ,'FIRST_NAME','SALARY']].groupby(['SALARY']).sum()
```
